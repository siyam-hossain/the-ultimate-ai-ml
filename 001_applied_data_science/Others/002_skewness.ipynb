```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import skew
import seaborn as sns
```

```python
# generating data

data1 = np.random.normal(
    loc = 0,
    scale = 1,
    size = 1000
)

skewness = skew(data1)
print("Skewness: ",skewness)
```

```python
sns.histplot(
    data = data1, 
    kde = True
)
```

```python
data2 = pd.DataFrame(
    {
        'values': np.random.lognormal(
            mean=0,
            sigma=1,
            size=1000
        )
    }
)

data2.head()
```

```python
skewness = skew(data2)
print(f"skewness: {skewness}")
```

```python
sns.histplot(
    data2['values'],
    kde=True
)
```

### **Right-skewed disribution**

```python
data = np.random.lognormal(
    mean=0,
    sigma=1,
    size=1000
)
```

```python
mean = np.mean(data)
median = np.median(data)
mode = max(
    set(data),
    key=list(data).count
)
```

```python
plt.figure(
    figsize=(20,5)
)

plt.hist(
    data, bins = 30, alpha = 0.6, color='blue', label='Data'
)

plt.axvline(
    mean, color='red', linestyle='dashed', linewidth='2', label = f"mean: {mean:.2f}"
)
plt.axvline(
    median, color='green', linestyle='dashed', linewidth='2', label = f"median: {median:.2f}"
)
plt.axvline(
    mode, color='red', linestyle='dashed', linewidth='2', label = f"mode: {mode:.2f}"
)

plt.title("Right-Skewed Distribution")
plt.xlabel('Values')
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
