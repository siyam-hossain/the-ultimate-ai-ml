```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(
    0, 
    4*np.pi,
    100
)
y = np.sin(x)
```

```python
print(x)
```

```python
print(y)
```

```python
plt.plot(
    x,
    y
)
plt.show()
```

```python
x = np.linspace(
    0, 
    2*np.pi,
    100
)
y = np.sin(x)

plt.plot(
    x,
    y
)
plt.axhline(
    y = 0,
    color = 'red',
    linestyle = '--',
    linewidth = 3
)
plt.show()
```

```python
x = np.arange(
    0, 
    2*np.pi,
    0.1
)
y = np.sin(x)

plt.plot(
    x,
    y
)
plt.axhline(
    y = 0,
    color = 'red',
    linestyle = '--',
    linewidth = 3
)
plt.show()
```
