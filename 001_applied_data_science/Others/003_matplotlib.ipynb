```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [3,6,7,9,20,4]

plt.scatter(
    x,
    y
)

plt.xlabel('X - axis')
plt.ylabel('Y - axis')
```

```python
import numpy as np

x = np.random.rand(80)
y = np.random.rand(80)


plt.scatter(
    x,
    y
)

plt.xlabel("NP X value")
plt.ylabel("NP Y value")
```

```python
# fixed the randomness

import numpy as np

np.random.seed(1)
x = np.random.rand(80)
y = np.random.rand(80)


plt.scatter(
    x,
    y
)

plt.xlabel("NP X value")
plt.ylabel("NP Y value")
plt.show()
```
