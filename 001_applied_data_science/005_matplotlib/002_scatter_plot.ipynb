### **Scatter plot**

+ bivariate analysis
+ numerical vs numerical
+ use case - finding correlation


**Simple function**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(-10,10,50)
x
```

```python
y = 10*x + 3 + np.random.randint(0,300,50)
y
```

```python
plt.scatter(
    x,
    y
)
```

### **scatter on pandas data**

```python
df = pd.read_csv('batter.csv')
df
```

```python
df = df.head(50)
```

```python
plt.scatter(
    df['avg'],
    df['strike_rate'],
    color = '#EE2835',
    marker= '+'
)

plt.title("Avg and SR analysis of top 50 Batsman")
plt.xlabel('Avg')
plt.ylabel('SR')
```

### **size**

```python
tips = sns.load_dataset('tips')
tips
```

```python
plt.scatter(
    tips['total_bill'],
    tips['tip'],
    
    
    s = tips['size']
)
```

```python
plt.scatter(
    tips['total_bill'],
    tips['tip'],
    s = tips['size']*20
)
```

### **Scatterplot using plt.plot**

+ faster technique (large dataset)
+ but less customization friendly 



```python
plt.plot(
    tips['total_bill'],
    tips['tip'],
    'o'
)
```
