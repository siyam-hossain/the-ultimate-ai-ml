## **Figure size, Annotations**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
iris['Species'] = iris['Species'].replace(
    {
        'Iris-setosa': 0,
        'Iris-versicolor': 1,
        'Iris-virginica': 2
    }
)
```

### **Figure size**

```python
plt.figure(
    figsize = (
        # width, height
        15, 7
    )
)


plt.scatter(
    iris['SepalLengthCm'],
    iris['PetalLengthCm'],
    
    c = iris['Species'],
    
    # to change color
    cmap = 'winter'
)


plt.colorbar()

plt.xlabel('Sepal-length')
plt.ylabel('Petal-length')

# plt.show()
```

### **Annotations**

```python
batters = pd.read_csv('batter.csv')
batters.head()
```

```python
sample_df = batters.head(100).sample(25, random_state=5)
sample_df
```

```python
x = [1,2,3,4]
y = [5,6,7,8]

plt.scatter(
    x,
    y
)

plt.text(1,5, 'Point-1', fontdict={'size':12, 'color':'red'})
plt.text(2,6, 'Point-2')
plt.text(3,7, 'Point-3')
plt.text(4,8, 'Point-4')
```

```python
plt.figure(
    figsize = (15, 8)
)


plt.scatter(
    sample_df['avg'],
    sample_df['strike_rate']
)

for i in range(sample_df.shape[0]):
    plt.text(
        sample_df['avg'].values[i], 
        sample_df['strike_rate'].values[i], 
        sample_df['batter'].values[i]
    )
```

### **dot size increment / decrement by batsman runs**

```python
plt.figure(
    figsize = (15, 8)
)


plt.scatter(
    sample_df['avg'],
    sample_df['strike_rate'],
    s = sample_df['runs']
)

for i in range(sample_df.shape[0]):
    plt.text(
        sample_df['avg'].values[i], 
        sample_df['strike_rate'].values[i], 
        sample_df['batter'].values[i]
    )
```

### **Horizontal and Vertical line**

```python
plt.figure(
    figsize = (15, 8)
)


plt.scatter(
    sample_df['avg'],
    sample_df['strike_rate'],
    s = sample_df['runs']
)

# strike rate 130
plt.axhline(
    130,
    color = 'red'
)
# avg 30
plt.axvline(
    30,
    color = 'red'
)

for i in range(sample_df.shape[0]):
    plt.text(
        sample_df['avg'].values[i], 
        sample_df['strike_rate'].values[i], 
        sample_df['batter'].values[i]
    )
```

```python
plt.figure(
    figsize = (15, 8)
)


plt.scatter(
    sample_df['avg'],
    sample_df['strike_rate'],
    s = sample_df['runs']
)

# strike rate 130
plt.axhline(
    130,
    color = 'red'
)
# avg 30
plt.axvline(
    30,
    color = 'red'
)
plt.axhline(
    140,
    color = 'green'
)

for i in range(sample_df.shape[0]):
    plt.text(
        sample_df['avg'].values[i], 
        sample_df['strike_rate'].values[i], 
        sample_df['batter'].values[i]
    )
```
