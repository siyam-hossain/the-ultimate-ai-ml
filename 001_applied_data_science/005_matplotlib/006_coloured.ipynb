```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('iris.csv')
iris.head()
```

```python
plt.scatter(
    iris['SepalLengthCm'],
    iris['PetalLengthCm']
)

plt.xlabel('Sepal-length')
plt.ylabel('Petal-length')

plt.show()
```

```python
iris.sample(5)
```

```python
iris['Species'] = iris['Species'].replace(
    {
        'Iris-setosa': 0,
        'Iris-versicolor': 1,
        'Iris-virginica': 2
    }
)
```

```python
iris.sample(5)
```

```python
plt.scatter(
    iris['SepalLengthCm'],
    iris['PetalLengthCm'],
    
    c = iris['Species'],
    
    # to change color
    cmap = 'winter'
)


plt.colorbar()

plt.xlabel('Sepal-length')
plt.ylabel('Petal-length')

plt.show()
```

```python
plt.scatter(
    iris['SepalLengthCm'],
    iris['PetalLengthCm'],
    
    c = iris['Species'],
    
    # to change color
    cmap = 'winter',
    
    # adding transparency
    alpha= 0.2
)


plt.colorbar()

plt.xlabel('Sepal-length')
plt.ylabel('Petal-length')

plt.show()
```
