### **From a pandas dataframe**

```python
batsman = pd.read_csv('sharma-kohli.csv')
batsman
```

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma']
)
```

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma']
)

plt.plot(
    batsman['index'],
    batsman['V Kohli']
)
```

### **Label Title**

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma']
)

plt.plot(
    batsman['index'],
    batsman['V Kohli']
)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### **colors(hex) and line(width and style) and marker(size)**

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma'],
    color = '#D3045D'
)

plt.plot(
    batsman['index'],
    batsman['V Kohli'],
    color = '#4750DD'
)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma'],
    color = '#D3045D',
    linestyle = 'dashed',
    linewidth=3
)

plt.plot(
    batsman['index'],
    batsman['V Kohli'],
    color = '#4750DD',
    linestyle = 'dotted',
    linewidth = 2
)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma'],
    color = '#D3045D',
    linestyle = 'dashed',
    linewidth=3,
    marker = 'o'
)

plt.plot(
    batsman['index'],
    batsman['V Kohli'],
    color = '#4750DD',
    linestyle = 'dotted',
    linewidth = 2,
    marker = 'D',
    markersize = 10
)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')
```

### **Legend: Location**

```python
plt.plot(
    batsman['index'],
    batsman['RG Sharma'],
    color = '#D3045D',
    linestyle = 'dashed',
    linewidth=3,
    marker = 'o',
    label = 'Rohit'
)

plt.plot(
    batsman['index'],
    batsman['V Kohli'],
    color = '#4750DD',
    linestyle = 'dotted',
    linewidth = 2,
    marker = 'D',
    markersize = 10,
    label = 'Virat'
)

plt.title("Rohit Sharma Vs Virat Kohli Career Comparison")
plt.xlabel('Season')
plt.ylabel('Runs Scored')

plt.legend(
    # loc = 'lower right'
    loc = 'best'
)
```

### **Limiting axes**

+ when outliers present in dataset

```python
price = [48000,54000,57000,49000,47000,45000, 4500000]
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(
    year,
    price
)
```

```python
price = [48000,54000,57000,49000,47000,45000, 4500000]
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(
    year,
    price
)

plt.ylim(
    0, 75000
)
```

```python
price = [48000,54000,57000,49000,47000,45000, 4500000]
year = [2015,2016,2017,2018,2019,2020, 2021]

plt.plot(
    year,
    price
)

plt.ylim(
    0, 75000
)

plt.xlim(
    2017,2019
)
```
