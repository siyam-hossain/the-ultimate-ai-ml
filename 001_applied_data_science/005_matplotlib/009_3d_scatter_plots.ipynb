```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_id = "1-sNYID0jjK5oW-KnsrGwZFh_H5C-Hqss"
url = f"https://drive.google.com/uc?id={file_id}"

url
```

```python
batters = pd.read_csv(
    url
)


batters.head()
```

```python
fig = plt.figure()

ax = plt.subplot(
    projection = '3d'
)
```

```python
fig = plt.figure(
    figsize=(8,8)
)

ax = plt.subplot(
    projection = '3d'
)

ax.scatter3D(
    batters['runs'],
    batters['avg'],
    batters['strike_rate'],
    color = 'red',
    edgecolor = 'black'
)
ax.set_title("IPL batsman analysis")
ax.set_xlabel('Runs')
ax.set_ylabel('Avg')
ax.set_zlabel('SR')
```

### **3D line plot**

```python
x = [0,1,5]
y = [0,10,13]
z = [0,13,20]
```

```python
fig = plt.figure(
    figsize=(8,8)
)

ax = plt.subplot(
    projection = '3d'
)

ax.scatter3D(
    x,
    y,
    z,
    color = 'red',
    edgecolor = 'black',
    s = [100,100,100] # for increasing size
)

ax.plot3D(
    x,
    y,
    z,
    color = 'orange'
)
```

```python
x = [0,1,5,25]
y = [0,10,13,0]
z = [0,13,20,9]

fig = plt.figure(
    figsize=(8,8)
)

ax = plt.subplot(
    projection = '3d'
)

ax.scatter3D(
    x,
    y,
    z,
    color = 'red',
    edgecolor = 'black',
    s = [100,100,100,100] # for increasing size
)

ax.plot3D(
    x,
    y,
    z,
    color = 'orange'
)
```
