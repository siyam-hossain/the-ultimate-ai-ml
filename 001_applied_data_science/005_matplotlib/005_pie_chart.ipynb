### **Pie Chart**

+ univariate / Bivariate analysis
+ categorical vs numerical 
+ use case: to find contribution on a standard scale

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### **simple data**

```python
data = [23,45,100,20,49]
subjects = ['eng', 'science', 'maths', 'sst', 'bangla']


plt.pie(
    data,
    labels = subjects
)

plt.show()
```

```python
df = pd.read_csv('gayle-175.csv')
df
```

### **Data set**

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%'
)

plt.show()
```

### **Specifing colors**

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    colors = ['blue', 'green', 'yellow', 'pink', 'cyan', 'brown']
)

plt.show()
```

### **Extracting a perticular section**

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0.1,0,0,0,0,0]

)

plt.show()
```

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0,0.1,0,0,0,0]

)

plt.show()
```

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0.1,0,0,0,0,0.3]

)

plt.show()
```

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0.1,0,0,0,0,0.3],
    shadow = True

)

plt.show()
```

## **Changing style**

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0.1,0,0,0,0,0.3]

)

plt.show()
```

## **Save figure**

```python
plt.pie(
    df['batsman_runs'],
    labels = df['batsman'],
    autopct = '%0.1f%%',
    
    explode = [0.1,0,0,0,0,0.3]

)

plt.savefig('sample-pie.png')
```
