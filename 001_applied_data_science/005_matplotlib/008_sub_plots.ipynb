```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_id = "1-sNYID0jjK5oW-KnsrGwZFh_H5C-Hqss"
url = f"https://drive.google.com/uc?id={file_id}"
```

```python
batters = pd.read_csv(
    url,
    sep=None,
    engine="python", 
    on_bad_lines="skip"
)

batters.head()
```

```python
plt.scatter(
    batters['avg'],
    batters['strike_rate']
)
```

### **Setting title; x, y label; figsize**

```python
fig, ax = plt.subplots(
    figsize = (8,6)
)

ax.scatter(
    batters['avg'],
    batters['strike_rate']
)

ax.set_title('Scatter Plot')
ax.set_xlabel('Avg')
ax.set_ylabel('SR')

plt.show()
```

```python
fig, ax = plt.subplots(
    nrows=2, 
    ncols=1
)

ax[0].scatter(
    batters['avg'],
    batters['strike_rate']
)
ax[0].set_title("Avg Vs SR")
ax[0].set_xlabel("Avg")
ax[0].set_ylabel("SR")

ax[1].scatter(
    batters['avg'],
    batters['runs']
)
ax[1].set_title("Avg Vs Runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")



plt.show()
```

```python
fig, ax = plt.subplots(
    nrows=2, 
    ncols=1,
    sharex=True,
    figsize = (10,5)
)

ax[0].scatter(
    batters['avg'],
    batters['strike_rate'],
    color = 'red'
)
ax[0].set_title("Avg Vs SR")
ax[0].set_ylabel("SR")

ax[1].scatter(
    batters['avg'],
    batters['runs']
)
ax[1].set_title("Avg Vs Runs")
ax[1].set_xlabel("Avg")
ax[1].set_ylabel("Runs")



plt.show()
```

### **Most Preferable Option**

```python
fig, ax = plt.subplots(
    nrows=2, 
    ncols=2,
    figsize = (10,5)
)

ax[0,0].scatter(
    batters['avg'], 
    batters['strike_rate'],
    color = 'red',
    edgecolor = 'black'
)
ax[0,1].scatter(
    batters['avg'], 
    batters['runs'],
    color = 'purple',
    edgecolor = 'black'
)
ax[1,0].hist(
    batters['avg'],
    color = 'red',
    edgecolor = 'black'
)
ax[1,1].hist(
    batters['runs'],
    color = 'purple',
    edgecolor = 'black'
)



plt.show()
```

### **Third Method**

```python
fig = plt.figure()

# 211 means 2 rows, 1 col, 1st graph
ax1 = fig.add_subplot(2, 1, 1)
ax1.scatter(
    batters['avg'],
    batters['strike_rate'],
    color = 'red',
    edgecolors = 'black'
)


# 211 means 2 rows, 1 col, 2nd graph
ax2 = fig.add_subplot(2, 1, 2)
ax2.scatter(
    batters['avg'],
    batters['runs'],
    color = 'purple',
    edgecolors = 'black'
)

plt.show()
```

```python
fig = plt.figure(
    figsize = (10,5)
)


ax1 = fig.add_subplot(2, 2, 1)
ax1.scatter(
    batters['avg'],
    batters['strike_rate'],
    color = 'red',
    edgecolors = 'black'
)



ax2 = fig.add_subplot(2, 2, 2)
ax2.scatter(
    batters['avg'],
    batters['runs'],
    color = 'purple',
    edgecolors = 'black'
)

ax3 = fig.add_subplot(2, 2, 3)
ax3.hist(
    batters['avg'],
    color = 'red',
    edgecolor = 'black'
)

ax3 = fig.add_subplot(2, 2, 4)
ax3.hist(
    batters['runs'],
    color = 'purple',
    edgecolor = 'black'
)

plt.show()
```
