```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Simple bar chart**

```python
children = [10,20,40,10,30]
colors = ['red','blue', 'green', 'yellow', 'pink']

plt.bar(
    colors, 
    children
)
```

```python
children = [10,20,40,10,30]
colors = ['red','blue', 'green', 'yellow', 'pink']

plt.bar(
    colors, 
    children,
    color = ['#D41414','#E19B8B','#00C588','#403963', '#310A69']
)
```

### **Horizontal bar chart**

```python
plt.barh(
    colors, 
    children,
    color = ['#D41414','#E19B8B','#00C588','#403963', '#310A69'],
    
    
)

plt.title('Children Vs Colors')
plt.xlabel('Children')
plt.ylabel('Colors')
```

### **Multiple Bar Charts**

```python
df = pd.read_csv('batsman_season_record.csv')

df.head()
```

```python
df.shape[0] # number of rows
```

```python
np.arange(df.shape[0])
```

```python
plt.bar(
    np.arange(df.shape[0]) - 0.2,
    df['2015'],
    width=0.2,
    color = '#E01305'
)
plt.bar(
    np.arange(df.shape[0]),
    df['2016'],
    width=0.2,
    color = '#EDB55E'
)
plt.bar(
    np.arange(df.shape[0]) + 0.2,
    df['2017'],
    width=0.2,
    color = '#7D58D9'
)



# xtricks(existing lebel, replace with replace values)
plt.xticks(
    np.arange(df.shape[0]),
    df['batsman']
)

plt.show()
```

### **Stacked Bar chart**

```python
df
```

```python
plt.bar(
    df['batsman'],
    df['2017'],
    color = '#541414',
    label = '2017'
)
plt.bar(
    df['batsman'],
    df['2016'],
    color = '#04529D',
    
    bottom = df['2017'],
    label = '2016'
    
)
plt.bar(
    df['batsman'],
    df['2015'],
    color = '#384166',
    
    bottom = (df['2016'] + df['2017']),
    label = '2015'
)

plt.legend()
plt.show()
```

```python
# Bottom structure


#               2017
#       2016    2016
# 2015  2015    2015


plt.bar(
    df['batsman'],
    df['2015'],
    color = '#541414',
    label = '2015',
    width=0.5
)
plt.bar(
    df['batsman'],
    df['2016'],
    color = '#04529D',
    
    bottom = df['2015'],
    label = '2016',
    width=0.5
    
)
plt.bar(
    df['batsman'],
    df['2017'],
    color = '#384166',
    
    bottom = (df['2015'] + df['2016']),
    label = '2015',
    width=0.5
)

plt.legend()
plt.show()
```
