### **Finds top 5 batsman who make most 6's**

```python
mask6 = delivery['batsman_runs']==6
mask4 = delivery['batsman_runs']==4

n_delivery = delivery[mask4]
new_delivery = delivery[mask6]
```

```python
new_delivery.shape[0]
```

```python
n_delivery.shape[0]
```

+ One column → SeriesGroupBy
> df.groupby('key')['col'].sum()

+ Multiple columns → DataFrameGroupBy
> df.groupby('key')[['col1', 'col2']].mean()

+ Same thing using .agg()
> df.groupby('key').agg(total=('col', 'sum'))


```python
new_delivery.groupby('batsman')['batsman_runs'].count()
```

```python
new_delivery.groupby('batsman')['batsman_runs'].count().sort_values(
    ascending=False
).head()
```

```python
def top_5_who_make_this(runs):
    mask = delivery['batsman_runs']==runs
    n_delivery = delivery[mask]
    
    return(
        n_delivery
        .groupby('batsman')['batsman_runs']
        .count()
        .sort_values(ascending=False)
        .head()
    )
```

```python
top_5_who_make_this(3)
```

```python
top_5_who_make_this(1)
```

```python
top_5_who_make_this(2)
```

```python
def anyone_who_made_top_most_runs_against_3_teams(name):
    b_name = delivery[
        delivery['batsman'] == name
    ]
    
    return(
        b_name
        .groupby('bowling_team')['batsman_runs']
        .sum()
        .sort_values(ascending=False)
        .head(3)
    )
```
