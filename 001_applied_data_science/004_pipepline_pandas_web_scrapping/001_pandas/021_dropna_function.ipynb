```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('titanic.csv')
data.head()
```

### **Dropping missing data using dropna()**

```python
data.shape
```

```python
data.dropna().head()
```

```python
data.dropna().shape[0]
```

```python
data.dropna(
    how='all' # all means drop rows where all of the attribute has no value
).head()
```

```python
# this block operated by axis = 0 means operate by row
data.dropna(
    how='all' # all means drop rows where all of the attribute has no value
).shape
```

```python
# this block operated by axis = 1 means operate by col
data.dropna(
    axis=1,
    how='all' # all means drop rows where all of the attribute has no value
).shape
```

```python
data.dropna(
    axis=1,
    how='any'
).shape
```
