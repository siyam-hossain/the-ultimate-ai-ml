```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('matches.csv')
data.head()
```

### **Masking**

#### **single condition**

```python
mask = data['city'] == 'Hyderabad'

data[mask].head()
```

```python
type(mask)
```

```python
data[mask].shape
```

```python
def get_city_match_count(city):
    mask = data['city']== city
    return data[mask].shape[0]
    
```

```python
get_city_match_count('Bangalore')
```

```python
get_city_match_count('Kolkata')
```

#### **multiple condition**

```python
mask1 = data['city'] == 'Hyderabad'
mask1
```

```python
mask2 = data['date'] > '2010-01-01'
mask2
```

```python
# while performing and operation on boolean value we must use bitwise & insted of and
data[
    mask1 & mask2
].head()
```

```python
data[
    mask1 & mask2
].shape[0]
```
