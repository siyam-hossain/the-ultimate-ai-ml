```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("manasgarg/ipl")

print("Path to dataset files:", path)
```

```python
import os

os.listdir(path)
```

```python
csv_path1 = os.path.join(path, 'deliveries.csv')
csv_path1
```

```python
csv_path2 = os.path.join(path, 'matches.csv')
csv_path2
```

```python
import pandas as pd

delivery = pd.read_csv(csv_path1)
match = pd.read_csv(csv_path2)
```

```python
delivery.head()
```

```python
match.head()
```

+ Merge
+ first column : second column
+ season       : batsman name

```python
# merge
delivery.merge(
    match,
    how='inner',
    left_on='match_id',
    right_on='id'
).head()
```

```python
new = delivery.merge(
    match,
    how='inner',
    left_on='match_id',
    right_on='id'
)

new.shape
```

```python
print(delivery.shape)
print(match.shape)
```

**Multi-index Series**

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum()
```

**sort values**

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum().sort_values(
    ascending=False
)
```

### **converting into dataframe**

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum().sort_values(
    ascending=False
).reset_index()
```

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum().sort_values(
    ascending=False
).reset_index().drop_duplicates(
    subset='season',
    keep='first'
)
```

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum().sort_values(
    ascending=False
).reset_index().drop_duplicates(
    subset='season',
    keep='first'
).sort_values(
    by='season',
    ascending=True
)
```

```python
new.groupby(
    ['season','batsman']
)['batsman_runs'].sum().sort_values(
    ascending=False
).reset_index().drop_duplicates(
    subset='season',
    keep='first'
).sort_values(
    by='season',
    ascending=True
)[
    ['season','batsman']
]
```
